```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import tree
import pandas as pd
```

```python
dataset = pd.read_csv('InformationGainData.csv')
```

```python
dataset
```

```python
dataset.drop(['RID'],axis=1,inplace=True)
```

```python
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
dataset['age']=le.fit_transform(dataset['age'])
dataset['income']=le.fit_transform(dataset['income'])
dataset['student']=le.fit_transform(dataset['student'])
dataset['credit_rating']=le.fit_transform(dataset['credit_rating'])
dataset['Class:buys computer']=le.fit_transform(dataset['Class:buys computer'])
```

```python
dataset
```

```python
X = dataset.drop(['Class:buys computer'],axis=1)
Y = dataset['Class:buys computer']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classifier = GaussianNB()
classifier.fit(X_train, Y_train)
```

```python
X_test
```

```python
X_test_GV=[[1,0,1,0]]
```

```python
Y_pred  =  classifier.predict(X_test_GV)
```

```python
print(Y_pred)
```

```python
Y_pred  =  classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(Y_test,Y_pred))
tn,fp,fn,tp=confusion_matrix(Y_test,Y_pred).ravel()
(tn,fp,fn,tp)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(Y_test,Y_pred)
accuracy
```
